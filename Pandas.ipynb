## A lot of mess, isn't it, let's sort it out

```python
def plot(X):
    try:
        data[X].plot()##add figsize = (20,14) to zoom into the picture
        plt.legend(X)
        plt.xlabel("Index")
        plt.ylabel(X)
        plt.show()
    except:
        pass
```

```python
# if you don't know how to use map function, please check out Part 2 of Tutorial 4
%matplotlib inline
print(list(map(plot, data.columns)))
```
